```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import talib as ta
```

```python
df = pd.read_csv('../../../data/stocks/AAPL.csv')
df
```

```python
df['SAR'] = ta.SAR(df['High'], df['Low'], acceleration=0.02, maximum=0.2)
df
```

```python
fig = go.Figure(
    go.Candlestick(
        x=df['Date'],
        open=df['Open'],
        high=df['High'],
        low=df['Low'],
        close=df['Close'],
        name='Price'
    )
)

fig.add_trace(
    go.Scatter(
        x=df['Date'],
        y=df['SAR'],
        mode='lines+markers',
        name='Parabolic SAR',
        marker=dict(color='red', size=5)
    )
)

fig.update_layout(
    title='Parabolic SAR',
    xaxis_rangeslider_visible=False,
    xaxis_title='Date',
    yaxis_title='Price',
    template='plotly_dark' 
)

fig.show()
```
