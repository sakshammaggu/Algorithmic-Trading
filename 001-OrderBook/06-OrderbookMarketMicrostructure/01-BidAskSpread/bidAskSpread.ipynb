```python
import pandas as pd
import requests
```

```python
url = f"https://api.binance.com/api/v3/depth"
params = {
    "symbol": "SOLUSDT",
    "limit": 10
}
```

```python
orderbook_data = requests.get(url, params=params).json()
orderbook_data
```

```python
ask_levels = pd.DataFrame(orderbook_data['asks'], columns=['price', 'quantity'], dtype=float)
ask_levels["side"] = "ask"
ask_levels = ask_levels.sort_values(by='price')
ask_levels
```

```python
bid_levels = pd.DataFrame(orderbook_data['bids'], columns=['price', 'quantity'], dtype=float)
bid_levels["side"] = "bid"
bid_levels = bid_levels.sort_values(by='price')
bid_levels
```

```python
orderbook = pd.concat([bid_levels, ask_levels]).sort_values(by="price", ascending=False)
orderbook
```

```python
best_ask = ask_levels['price'].min()
best_bid = bid_levels['price'].max()
bid_ask_spread = best_ask - best_bid
print(f"Best Ask: {best_ask} \nBest Bid: {best_bid} \nBid-Ask Spread: {bid_ask_spread}")
```
